```python
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
```

# Определение заданной функции

```python
def math_func(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

Найдем значения функции в точках x = 1 и x = 15 и составим СЛАУ для многочлена первой степени:

```python
x = [1, 15]
f = [math_func(x[0]), math_func(x[1])]
A = [[1, x[0]], [1, x[1]]]
b = f
w = linalg.solve(A, b)

N = 1000
X = np.linspace(1, 15, N, dtype=float)
F = np.zeros(N)
for i in range(N):
    F[i] = math_func(X[i])
F_app1 = w[0] + w[1] * X
```

```python
plt.figure(figsize=(16, 9))
plt.plot(X, F)
plt.plot(X, F_app1)
```

Из графиков можно сделать вывод, что многочлен первой степени плохо приближает вектор значений к истинной зависимости.
Следовательно, необходимо пробывать подбирать многочлен более высокого порядка.
Найдем значения функции в точках x = 1, x = 8 и x = 15 и составим СЛАУ для многочлена второй степени:

```python
x = [1, 8, 15]
f = [math_func(x[0]), math_func(x[1]), math_func(x[2])]
A = [[1, x[0], x[0] ** 2], [1, x[1], x[2] ** 2], [1, x[2], x[2] ** 2]]
b = f
w = linalg.solve(A, b)
F_app2 = w[0] + w[1] * X + w[2] * X ** 2
```

```python
plt.figure(figsize=(16, 9))
plt.plot(X, F)
plt.plot(X, F_app2)
```

Многочлен второго порядка также дает плохое приближения.
Повторим итерацию для многочлена 3 степени и значений функций в точках x = 1, x = 4, x = 10 и x = 15:

```python
x = [1, 4, 10, 15]
f = [math_func(x[0]), math_func(x[1]), math_func(x[2]), math_func(x[3])]
A = [[1, x[0], x[0] ** 2, x[0] ** 3], [1, x[1], x[1] ** 2, x[1] ** 3], [1, x[2], x[2] ** 2, x[2] ** 3], [1, x[3], x[3] ** 2, x[3] ** 3]]
b = f
w = linalg.solve(A, b)
F_app3 = w[0] + w[1] * X + w[2] * X ** 2 + w[3] * X ** 3
```

```python
plt.figure(figsize=(16, 9))
plt.plot(X, F)
plt.plot(X, F_app3)
```

Многочлен 3 степени достаточно хорошо приближает точки к реальной зависимости по сравнению с первой и второй степенями.

Для наглядности, отобразим реальную зависимости и многочлены первого, второго и третьего порядка на одном графике.

```python
plt.figure(figsize=(16, 9))
plt.grid()
plt.plot(X, F)
plt.plot(X, F_app1)
plt.plot(X, F_app2)
plt.plot(X, F_app3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Исконная зависимость', 'Многочлен 1 степени', 'Многочлен 2 степени', 'Многочлен 3 степени'])
```
