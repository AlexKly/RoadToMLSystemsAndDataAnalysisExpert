# Центральная предлельная теорема (ЦПТ) своими руками

```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

# Бета-распределение непрерывной величины

Описание Бета-распределения: https://ru.wikipedia.org/wiki/Бета-распределение

Среднее:
$E[x]=\frac{\alpha}{\alpha+\beta}$

Дисперсия:
$D[x]=\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$


```python
# Задаем параметры распределения:
a, b = 0.5, 0.5

# Подсчитаем матожидание и дисперсию:
E = a / (a + b)
D = a * b / ((a + b) ** 2 * (a + b + 1))
print('Матожидание:', E)
print('Дисперсия:', D)
```

```python
# Генерируем выборку объема 1000:
samples = sts.beta.rvs(a, b, size=1000) # Генерируем случайную выборку объемом 1000 значений
x = np.linspace(0, 1, 1000)
pdf_beta = sts.beta.pdf(x, a, b) # Плотность распределения
```

```python
# Построение гистограммы выборки и теоретической плотности распределения случайной величины:
plt.figure(figsize=(15,5))
plt.title('Betta-distribution', fontdict={'fontsize': 18})
plt.plot(x, pdf_beta, color='red', label='Theoretical PDF')
plt.hist(samples, bins=40, range=(0,1), density=True, color='blue', label='Normalized histogram of sample distribution')
plt.ylabel('Normalized number of counts' ,fontdict={'fontsize': 14})
plt.xlabel('Random variable value', fontdict={'fontsize': 14})
plt.legend()
plt.show()
```

# Оценка распределения выборочного среднего случайной величины при разных объёмах выборок (n)

## n = 5:

```python
n = 5
samples_n1 = np.array([]) # Массив выборочных средних

for i in range(1000): # Цикл генерации случайных выборок количества n из функции
    samples_n1i = sts.beta.rvs(a, b, size=n)
    mean_n1i = sum(samples_n1i)/n # Выборочное среднее
    samples_n1 = np.append(samples_n1, mean_n1i) # Массив из выборочных средних


D_n1 = D/n # Дисперсия для нормального распределения
sigma_n1 = math.sqrt(D_n1) # СКО для нормального распределения
norm_n1 = sts.norm(E, sigma_n1) # Нормальное распределение с расчетными характеристиками Бета-распределения
pdf_n1 = norm_n1.pdf(x) # Плотность распределения
```

```python
# Гистограмма средних и плотность нормального распределения
plt.figure(figsize=(15,5))
plt.title('Distribution of sample means for 1000 samples volume of 5', fontdict={'fontsize': 15})
plt.plot(x, pdf_n1, color='red', label='PDF (n = 5)')
plt.hist(samples_n1, bins=40, range=(0,1), density=True, color='blue', label='Normalized histogram of mean distribution')
plt.ylabel('Normalized number of counts' ,fontdict={'fontsize': 14})
plt.xlabel('Sample mean value', fontdict={'fontsize': 14})
plt.legend(loc='best')
plt.show()
```
